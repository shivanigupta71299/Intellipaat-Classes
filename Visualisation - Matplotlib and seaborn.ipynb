```python
# this should be done before plot otherwise plot nhi hoga. it is graph size kya hona chahiye
# Resizing the graph
plt.figure(figsize=(5,5),dpi=100)

plt.plot(y,x,label='2*n',color='#D9DF28',marker='*',linewidth=2,markersize=10,markeredgecolor='black',linestyle=':')  
#linestyle - 
'''
'-' for straight line
':' for dotted line
'--' for dasshed line

'''
# Marker is used to highlight the point of plot
plt.title('Testing the data',fontdict={'fontname':'Times new Roman'})
plt.title('Testing the data',fontdict={'fontname':'Consolas','fontsize':'20'})
plt.xlabel('Test Data X -Axis',fontdict={'fontname':'Consolas','fontsize':'15'})
plt.ylabel('Test Data Y -Axis')

# To define values to take on X-axis
# plt.xticks([2,4,10,5])
plt.xticks(range(1,10))

# To define values to take on Y-axis
# plt.yticks([1,2,4,10,15,3,5])
plt.yticks(range(2,70,5))  # skip alternate and show bw range 1 to 14


# Line 2
x2=np.array([1,2,3,4])
plt.plot(x2,x2**3,label='3*n',marker='^',markeredgecolor='black',linewidth=2,linestyle='--',color='red')


# Line 3
x3=np.array([2,4,6,8,10])
plt.plot(x3,x3/2,label='n/2',marker='^',markeredgecolor='black',linewidth=2,linestyle='-',color='violet')

# For Legend(dictionary- ki konsi color ka kya code h.)  - add label in plot statement
# should be at end
plt.legend()

#save the graph
plt.savefig('FirstGraph.png',dpi=100)

plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(5,4),dpi=100)

x=['a','b','c','d','e']
y=[1,5,3,4,6]
bar_chart =plt.bar(x,y)
print(bar_chart)
print(bar_chart[0])
bar_chart[0].set_hatch('.')
bar_chart[1].set_hatch('/')
bar_chart[2].set_hatch('*')
bar_chart[3].set_hatch('^')
bar_chart[4].set_hatch('.')

plt.title('Bar Chart')
plt.xlabel('CHARACTER')
plt.ylabel('COUNT')

plt.show()
```

## 1. Create scatter plots for the following data, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced. 
a. Create scatter plots with between (mpg-hp, hp-weight), and (hp-mpg, hp-time-to-60) with respect to each brand.



```python
# Use HUE for brand filter
plt.subplots(2,2,figsize=(15,15))

plt.subplot(2,2,1)
plt.scatter(x= df['hp'],y=df['mpg'])

plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.show()
```

## 2. Create bar plots for the following, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.
a. Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands.

```python
# x- cyl, y-  hp, mpg, weight, time-to-60  one by one


plt.subplots(2,2,figsize=(10,5))

plt.subplot(2,2,1)
sns.barplot(x=df['cylinders'],y=df['hp'],hue=df['brand'])
plt.title('cylinders vs hp')

plt.subplot(2,2,2)
sns.barplot(x=df['cylinders'],y=df['mpg'],hue=df['brand'])
plt.title('cylinders vs mpg')

plt.subplot(2,2,3)
sns.barplot(x=df['cylinders'],y=df['weightlbs'],hue=df['brand'])
plt.title('cylinders vs weightlbs')

plt.subplot(2,2,4)
sns.barplot(x=df['cylinders'],y=df['time-to-60'],hue=df['brand'])
plt.title('cylinders vs time-to-60')

plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.show()
```

b. Create a bar plot that shows the visual representation of hp, mpg, weight and time-to-60 with respect to the years, for each brand.

```python
# x- cyl, y-  hp, mpg, weight, time-to-60  one by one


plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.barplot(x=df['year'],y=df['hp'])
plt.title('year vs hp')

plt.subplot(2,2,2)
sns.barplot(x=df['year'],y=df['mpg'],hue=df['brand'])
plt.title('year vs mpg')

plt.subplot(2,2,3)
sns.barplot(x=df['year'],y=df['weightlbs'],hue=df['brand'])
plt.title('year vs weightlbs')

plt.subplot(2,2,4)
sns.barplot(x=df['year'],y=df['time-to-60'],hue=df['brand'])
plt.title('year vs time-to-60')

plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.show()
```
