```python
A=pd.DataFrame({
    'color':['red','yellow','green'],
    'countofshirt': [1,2,3],
    
})
B=pd.DataFrame({
    'color':['yellow','violet','green'],
    'Shirtsize': ['S','M','L'],
    
})
```

```python
pd.merge(A,B,how='inner')
```

```python
 ## Outer Join Or Full Join
# select * from A join B on A.color=B.color
pd.merge(A,B,how='outer')
```

```python
## left join
pd.merge(A,B,how='left')
pd.merge(B,A,how='left')
```

```python
## Right join
pd.merge(A,B,how='right')
```

```python
A=pd.DataFrame({
    'color':['red','yellow','green'],
    'countofshirt': [1,2,3],
    
})
B=pd.DataFrame({
    't-shirt':['yellow','violet','green'],
    'Shirtsize': ['S','M','L'],
    
})
```

```python
# select * from A join B on A.color=B.t-shirt
pd.merge(A,B,left_on='color',right_on='t-shirt')
```

```python
pd.merge(A,B,left_on='color',right_on='t-shirt', how='outer')
```

```python
pd.merge(A,B,left_on='color',right_on='t-shirt', how='right')
```
