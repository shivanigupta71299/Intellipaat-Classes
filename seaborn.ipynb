```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# steps to read and visualise the data

'''
 1.Import Libraries
 2.Load the dataset using pandas.
 3. Explore the pandas dataframe using the head() and info() and shape functions.
 4.Visualize the Data - eg. we can use a scatter plot using Seaborn to visualize the data, 
 5. Clean the data
 
'''
```

```python
df = pd.read_csv('./mtcars.csv')
df.head()
df.shape
```

```python
# barplot
sns.barplot(x = df['cyl'], y = df['mpg'])
plt.show()
```

```python
sns.barplot(x = df['cyl'], y = df['wt'])
plt.show()
```

```python
sns.barplot(x='cyl', y='wt', data=df)
plt.show()
```

```python
sns.barplot(x='cyl', y='wt', data=df, color='yellow')
plt.show()
```

```python
# color palettes -> theme
# sns.barplot(x='cyl', y='wt', data=df, palette='Set1')
# sns.barplot(x='cyl', y='wt', data=df, palette='Set2')
# sns.barplot(x='cyl', y='wt', data=df, palette='rocket')
plt.show()
```

```python
# count plot -> 1 axis as input and other -> count
sns.countplot(x='cyl', data=df, palette='Set2')
plt.show()
```

```python
sns.countplot(x='gear', data=df, palette='Set2')
plt.show()
```

```python
sns.countplot(y='gear', data=df, palette='Set2')
plt.show()
```

```python
sns.countplot(x='cyl', y='gear', data=df, palette='Set2') # error
plt.show()
```

```python
# heatmap -> correlation 
# sns.heatmap(df.corr(), cbar=False)
sns.heatmap(df.corr(), cbar=True, linewidths=1)
plt.show()
```

```python
# scatterplot
sns.scatterplot(x='disp', y='hp', data=df)
plt.show()
```

```python
sns.scatterplot(x='disp', y='hp', data=df, hue='mpg')
plt.show()
```

```python
# pairplot
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df, hue='mpg', palette="Set2")
plt.show()
```
