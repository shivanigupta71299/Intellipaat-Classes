```python
data['start'] = data['Span'].apply(lambda x: int(x.split('-')[0]))
data['end'] = data['Span'].apply(lambda x: int(x.split('-')[1]))

data.head()
```

```python
data['Exp'] = data['end']-data['start']
data.head()
```

```python
data['HS'] = data['HS'].str.extract('(\d+)').astype('int')
```

```python
# outlier treatment

features = [ 'Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', 'Exp' ]

for feature in features:

    p99 = data[feature].quantile(0.99)
    p1 = data[feature].quantile(0.01)

    data[feature][data[feature]>p99] = p99
    data[feature][data[feature]<p1] = p1
    
```

```python
clusters = list(range(2,8))
sse = [] # inertia

for num_clust in clusters:
    model_clus = KMeans(n_clusters=num_clust,
                       random_state=100)
    model_clus.fit(df_scaled)
    sse.append(model_clus.inertia_)

plt.plot(clusters, sse)
```

```python
clusters
```

```python
from sklearn.metrics import silhouette_score

clusters = list(range(2,8))

for num_clust in clusters:
    model_clus = KMeans(n_clusters=num_clust,
                       random_state=100)
    model_clus.fit(df_scaled)
    score = silhouette_score(df_scaled, model_clus.labels_)
    print('number of clusters = ',num_clust, ' : silhouette_score = ',score)
```

```python
# heirarchical clustering
from scipy.cluster.hierarchy import linkage, dendrogram, cut_tree
```

```python
df_scaled
```

```python
plt.figure(figsize=(20,10))
linkages = linkage(df_scaled, method='single', metric='euclidean')
dendrogram(linkages)
plt.show()
```

```python
plt.figure(figsize=(20,10))
linkages = linkage(df_scaled, method='average', metric='euclidean')
dendrogram(linkages)
plt.show()
```

```python
plt.figure(figsize=(20,10))
linkages = linkage(df_scaled, method='complete', metric='euclidean')
dendrogram(linkages)
plt.show()
```

```python
cluster_labels = cut_tree(linkages, n_clusters=2).reshape(79,)
cluster_labels
```

```python
pd.Series(cluster_labels).value_counts()
```

```python
data['cluster_id_hc'] = cluster_labels
data.head()
```
