```python
num = pd.Series([1,2,3,4])
print(num)
print(type(num))
for i in num:
    print(i)
```

```python
# another way to set Series
li=[1,2,3,4]
nums_series=pd.Series(li)
print(nums_series)
```

```python
num = pd.Series([1,2,3,4], index=['a','b','c','d'])
print(num)
# we can access data from initial and new indexes both
print(num[0],num['a'])
```

```python
# Series from dictionary

a={
    'name':'shivani',
    'age':'22',
    'age':'22'
}
print(a)

b=pd.Series(a)
print(b)
print(type(b))

print(b['age'])
```

```python
# Series from dictionary

a={
    'name':'shivani',
    'age':'22',
    'Mother':{'name':'Meenu',
    'age':'46',}
}
print(a)

b=pd.Series(a)
print(b)
# print(type(b))
print(b['Mother'])
print(b['Mother']['name'])
```

# Method 2 of data cleaning

```python
df=pd.read_csv('mtcars.csv')
df
```

```python
df.fillna(method='ffill')  # forward fill.
# for NaN, it replaces it by its preceding values
```

```python
df.fillna(method='ffill',limit=2)  # If 3 consecutive NaN, then it will fill only 2 by its preceding val
```

```python
df.fillna(method='bfill',limit=2)  # Backward Fill
#If 3 consecutive NaN, then it will fill only 2 by its next val
```

```python
df.interpolate()  # (next val - prev val)/number of NaN ---- It adds this number to previous val.

# When to use - When numbers are in increasing or decreasing pattern
```

```python
# Case - Dont include NaN vals. Drop those rows
df.dropna()# see 5-12 data is missing
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=3)  # if there are more than 3 values present then only include this row. If all are NaN then it will remove that row
#or it has 1 val and rest are NaN, it will drop that row
```

```python
df.replace('NaN',0)  # replace all nan by 0
```

```python
df.replace([0,1],100) # everywhere there is 0 and 1 it will replace }with 100
```
